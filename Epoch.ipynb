```python
%%writefile epoch.c

#include <stdio.h>
#include <stdlib.h>
#include <math.h>
#include <time.h>

#define TRUE 1
#define FALSE 0

// Tasa de aprendizaje
#define ALPHA 0.1

// Criterio de paro 
#define STOP_CRITERIA 100

int **crearPatrones(int rows, int n);
void imprimirTabla(int **patrons, int rows, int n);
void calcularPesos(float *weight, float bias, int **patrons, int *values, int n);
int evaluarPatrones(int *patrons, float *weight, float bias, int n);
float *modificarPesos(float *weight, int *patrons, int result, int n);
float modificarBias(float bias, int result);
float *pesosAleatorios(int size);
float biasAleatorio();

int main(void){

	int n = 2, rows = 4;
	
	// Generar valores de tabla de verdad
	int **patrons = crearPatrones(rows, n);
	  	
  // Creamos valores de salida
	int values[4] = {0,1,1,0};
  	
  // Pesos iniciales
	float *weight = (float *)malloc(sizeof(float) * 2);
	weight[0] = 1.0;
	weight[1] = 1.0;
	
	float bias = -0.5;
	
	weight = pesosAleatorios(2);
	bias = biasAleatorio();
		
  calcularPesos(weight, bias, patrons, values, rows);
  	
	return 0;
}

void calcularPesos(float *weight, float bias, int **patrons, int *values, int rows) {
	
	int i, j, epoch, err, result, real;
	epoch = err = 0;
	
	do{
		printf("\nEpoca %d\n", epoch);
		for(i = 0, err = 0; i < rows; i++){
			real = evaluarPatrones(patrons[i], weight, bias, rows);
			//printf("Valor esperado [%d] y Real [%d]\n", values[i], real);
			result = values[i] - real;
			//printf("Resultado[%d]: %d\n", i, result);
			if(result != 0){
				//printf("Entra[%i] con - %d -\n", i, result);
				err++;
				weight = modificarPesos(weight, patrons[i], result, rows);
				bias = modificarBias(bias, result);
				printf("Error: w1 = %f - w2 = %f - bias = %f\n", weight[0], weight[1], bias);
			}
		}
		if(err == 0){
			printf("Sin errores\n");
			printf("Pesos finales [w1 = %f | w2 = %f | bias = %f]\n", weight[0], weight[1], bias);
		}
		epoch++;
	}while(err != 0 && epoch < STOP_CRITERIA);

	if(epoch == STOP_CRITERIA - 1) {
		printf("No se encontraron pesos");
	}
	
}

float *modificarPesos(float *weight, int *patrons, int result, int n) {
	
	int i;
	
	for(i = 0; i < n; i++){
		weight[i] = weight[i] + (ALPHA * result * patrons[i]);
	}
	
	return weight;
	
}

float modificarBias(float bias, int result) {
	
	return bias + (ALPHA * result * 1);

}

float *pesosAleatorios(int size) {
	
	int i = 0;
	float *nWeight = (float *)malloc(sizeof(float) * size);
	float numero;
	
	for(i = 0; i < size; i++){
		numero = rand() % 101;
		
		nWeight[i] = numero / 100; 
	}
	
	return nWeight;
}

float biasAleatorio() {
	
	float numero;
	numero = rand() % 101;
	
	return numero / 100;
}

int evaluarPatrones(int *patrons, float *weight, float bias, int n) {
  
  	int i, j, val;
  	float sum;
  
  	for(i = 0, sum = 0; i < n; i++){
  		sum += patrons[i] * weight[i]; 
	}
	
	sum += bias;
	
	if(sum > 0){
		return 1;
	}else{
		return 0;
	}

}

void imprimirTabla(int **patrons, int rows, int n) {
	
	int i, j;
	
	// Generamos matriz con rows valores
	for (i = 0; i < rows; i++) {
	    for (j = 0; j < n; j++) {
		    printf("%d ", patrons[i][j]);
	    }
	    printf("\n");
	}
		
}

int **crearPatrones(int rows, int n) {
	
	int i, j; 
  	
  // Matriz de valores
  int **patrons = (int **)malloc(sizeof(int *) * rows);
  	
	for(i=0;i< rows;i++){
   	patrons[i] = (int*)malloc(sizeof(int) * n);
  } 
  	
  	// Llenar matriz
  	for (i = 0; i < rows; i++) {
	    int decimal = i;
	    for (j = n - 1; j >= 0; j--) {
		    int bit = decimal % 2;
	    	patrons[i][j] = bit;
	      	decimal /= 2;
    	}
  	}
	
	return patrons;
	
}
```
